# Introduction to Ggplot and Scatter Plots
`plot()` and numerous other related functions in the base distribution of R can be used to make beautiful and sophisticated plots. However, `ggplot` is a very popular function in data science for easily creating and customizing any type of plot. Conveniently, `ggplot` is located in the `ggplot2` package that is contained within the `tidyverse` package.

We use the following format for plotting data with ggplot: `ggplot(data = <Data_Frame>)+ GEOM_FUNCTION>(mapping=aes(<mappings>))`. `geom_function` refers to the type of graph we want to create. Since we will create scatter plot in this section, we will need to use `geom_point`, We will come to that later  but if we want to create bar plots, then we will use `geom_bar()` and so on. So if we want to plot previous graph with `ggplot` where we examine the relationship of highway miles per gallon as a function of engine displacement,we need to write the following code: `ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy))` 
In general the ggplot() function creates a coordinate system that we can add layers to, and we have told it use `mpg` to make subsequent layers. We then add a scatterplot layer using the `geom_point()` function, which takes the `mapping` argument. The `mapping` argument defines how variables are mapped to visual properties.The `mapping()` argument is always paired with the `aes()` function, along with its x and y arguments.The `aes()` function specifies which variables to map to the `x` and `y` axes, which happen to be `displ` and `hwy` respectively.

```r
ggplot(data=mpg) + geom_point(mapping=aes(x=displ,y=hwy))
```

**Assigning different colors based on car type:** We need to type `color=class` inside `aes()` function

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy,color=class))
```

**Assigning different sizes based on car type:** Although it is recommendable to use size for a discrete variable, we can assign that by typing `size=class` inside the `aes()` function.

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy,size=class))
```

**Assigning different transparencies based on car type:** we can type `alpha=class` in `aes()` function. Also we can use numbers as well to assign trasparency such as `alpha=0.1`.

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy,alpha=class),size=3,color="orange")
```

**Assigning different shapes based on car type:** we can type `shape=class` inside `aes()` function. This will create a diferent shape for each of the classes type.

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy,shape=class),size=3)
```

**Assigning the same color to all points:** If we want to assign anything to all data frame, we should not type it inside `aes()` function. HHere the color does not convey information about a feature, but only changes the appearance of the plot. Thats why we will be using `color="blue"` outside the `aes()` function since we are setting the color to blue as an argument of the `geom_point()` function and not the `aes()` function.

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy),color="blue")
```

**More Complicated Plots:** Now we will be using all the methods mentioned before in one plot.

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy,shape=drv,color=class),size=3)
```

**Using `jitter` to deal with overplotting:** The `mpg` dataset has 235 observations (rows), yet only 126 points exist in our plots.This problem is known as
**overplotting** , and occurs here because values of `displ` are rounded to one decimal place and values of hwy are rounded to the nearest integer.To deal with that problem we type `position="jitter"` in `geom_point()` function. Here,`jitter` is telling `geom_point()` to add some random noise to the `x` and `y` axis values of each point prior to it being plotted. We can also use automatic `jittering` with `geom_jitter()`

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy),position="jitter")
```

```r
ggplot(data=mpg)+geom_jitter(mapping=aes(x=displ,y=hwy))
```

**Facets:** Facets create subplots for different subsets of the data and are particularly useful for categorical data, where each subplot represents a distinct category. To facet a plot by a single variable (e.g.,car type), we can use the `facet_wrap()` function, where the first argument is a formula created using the tilde (`~`) character. We can also specify number of the rows with `nrow=2` or number of the columns with `ncol=2`

```r
ggplot(data=mpg)+geom_jitter(mapping=aes(x=displ,y=hwy))+facet_wrap(~class,nrow=2)
```

```r
ggplot(data=mpg)+geom_jitter(mapping=aes(x=displ,y=hwy))+facet_wrap(~class,ncol=2)
```

**Faceting data across a combination of two features:** `facet_wrap()` function allows us to split data based on a single feature. However,it can sometimes be useful to examine two features at once. The `facet_grid()` function allows the data to be plotted in subplots, with one feature representing subplot rows and another feature representing subplot columns. The first argument of facet_grid is a formula of the form: `row_feature ~ column_feature`. In out graph we used `drv~cyl` which means the columns which are `4,5,6,8` represent `drv` and rows which are `4,f,r` represent `cyl`

```r
ggplot(data=mpg)+geom_jitter(mapping=aes(x=displ,y=hwy))+facet_grid(drv~cyl)
```

**Plotting a smoothed line fitted with a confidence interval:** For ploting a smoothed line with a confidence interval we use `geom_smooth()` function.The format is very similar to `geom_point()`: `geom_smooth(mapping = aes(x=displ,y=hwy))`. We told R to plot a smoothed line and to display bands representing
95% confidence intervals (the default) around the smoothed line. If we want to change the confidence interval to 0.99 we need to type `level=0.99` in `geom_smooth()` function.

```r
ggplot(mpg)+geom_smooth(mapping=aes(x=displ,y=hwy))
```

```r
ggplot(mpg)+geom_smooth(mapping=aes(x=displ,y=hwy),level=0.99) #changing confidence interval to 0.99
```

**Plotting with separate fit and color for each drive type:** We need to type `color=drv` argument in  the `aes()` function.

```r
ggplot(mpg)+geom_smooth(mapping=aes(x=displ,y=hwy,color=drv))
```

**Adding layers of different geom plot types:** We have examined two geom plot types so far, which are the `geom_point()` for scatterplots and the `geom_smooth()` for smoothed line fits with confidence bands. Recall that ggplot simply initiates a blank plot, whereas `geom_point()` and `geom_smooth()` add layers to the plots. Therefore, we can add multiple layers to the same plot to make arbitrarily complex graphs. For example, adding a smoothed fit to a scatterplot permits visualization of overall trends while maintaining transparency about the underlying data.

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy))+geom_smooth(mapping=aes(x=displ,y=hwy))
```

```r
#Displaying points and fitted curve on same graph (compact)
ggplot(data=mpg,mapping=aes(x=displ,y=hwy))+geom_point()+geom_smooth()
```

```r
#Adding color to the scatterplot based on car type
ggplot(data=mpg,mapping=aes(x=displ,y=hwy))+geom_point(mapping=aes(color=class),size=3)+geom_smooth()
```

```r
#Adding color based on drive in scatter and smoothed plots
ggplot(data=mpg,mapping=aes(x=displ,y=hwy,color=drv))+geom_point()+geom_smooth()
```

# Bar Plots

Bar plots display frequencies of classes of categorical data. To explore the use of bar charts, we will examine a new data frame (tibble ) called `diamonds` which comes packaged with `tidyverse`. This data frame has 53,940 rows (observations) with 10 columns (features) containing attributes of diamonds.To be able to obtain descriptions of the features in `diamonds` ,we can type either `help(diamond)` or `?mpg`


```r
help(diamonds)
```

**Bar plot showing distribution of diamond quality:** We can display diamond quality (`cut`) on the `𝑥` axis, and the
frequency (count) for a diamond of that quality on the `𝑦` axis:

```r
ggplot(data = diamonds)+geom_bar(mapping=aes(x=cut))
```

**Displaying proportion rather than count (correct way):** We can also use `stat(prop)` to use the proportion, rather than count, as the frequency on the `𝑦` axis by setting the argument `group = 1`.

```r
ggplot(data = diamonds)+geom_bar(mapping=aes(x=cut,y=stat(prop),group=1))
```

```r
#Coloring bars based on category
ggplot(data = diamonds)+geom_bar(mapping=aes(x=cut,fill=cut))
```

```r
#Creating stacked bar charts showing different classes
ggplot(data = diamonds)+geom_bar(mapping=aes(x=cut,fill=clarity))
```

```r
#Creating stacked bar charts showing relative proportions
ggplot(data = diamonds)+geom_bar(mapping=aes(x=cut,fill=clarity), position="fill")
```

```r
#Creating side by side rather than stacked bar charts
ggplot(data = diamonds)+geom_bar(mapping=aes(x=cut,fill=clarity), position="dodge")
```

# Histograms

Histograms display distributions of continuous (quantitative) variables. For example, we can create a histogram showing the distribution of highway miles per gallon in the mpg data frame. If we do not state the `bin` number, `geom_bar()` will take it as 30.

```r
ggplot(data = mpg,mapping = aes (x = hwy))+geom_histogram(fill="orange")
```

```r
#Adjusting the number of bins creates smoother histogram
ggplot(data = mpg,mapping = aes (x = hwy))+geom_histogram(bins=10,fill="orange")
```

```r
#Similar effect when carefully adjusting bin width
ggplot(data = mpg,mapping = aes (x = hwy))+geom_histogram(binwidth=3.55,fill="orange")
```

# Boxplots

Boxplots are a compact way to summarize the distribution of points.


```r
#Creating boxplots
ggplot(data=mpg,mapping=aes(x=class,y=hwy))+geom_boxplot()
```

```r
#Easier visualization by flipping coordinates
ggplot(data=mpg,mapping=aes(x=class,y=hwy))+geom_boxplot()+coord_flip()
```

**Plotting objects can be assigned to a variable:** Here we created a bar plot with bars filled in colors corresponding to their `cut` , no legend for colors (`show.legend = FALSE` ), and no spacing between bars ( `width = 1`). We also made the width and height of the plots equal (`theme(aspect.ratio = 1)`) and removed `𝑥` and `𝑦` axis labels (`labs(x = NULL , y = NULL)`)

```r
bar=ggplot(data = diamonds)+ geom_bar(mapping = aes (x = cut, fill =cut),show.legend = FALSE , width = 1)+theme(aspect.ratio = 1)+theme(aspect.ratio = 1)
```

```r
#Typing this variable creates a bar plot with these characteristics:
bar
```

```r
#Can also manipulate the variable (e.g., flip coordinateS)
bar+coord_flip()
```

#  Coxcomb Chart

A Coxcomb chart is similar to a pie chart, except that angles are identical across categories and differences between categories are visualized based on the distances sectors extend from the center. We can easily switch to Coxcomb chart with polar coordinates with `coord_polar()`

```r
bar + coord_polar()
```

```r
#Change theme to black and white using theme_bw()
bar + coord_polar()+theme_bw()
```
